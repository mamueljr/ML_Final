```python
# estadisticas de tiros por jugador
tiros = pd.read_html('https://fbref.com/en/comps/12/shooting/La-Liga-Stats')
```

```python
for idx,table in enumerate(tiros):
    print("********************")
    print(idx)
    print(table)
```

```python
tiros = pd.DataFrame(tiros[0])
```

```python
import matplotlib.pyplot as plt
```

```python
tiros.columns
```

```python
tiros.columns = tiros.columns.droplevel()
```

```python
top4 = ['Barcelona', 'Atlético Madrid', 'Real Madrid', 'Real Sociedad']
```

```python
sub_set_tiros= tiros[tiros['Squad'].isin(top4)]
```

```python
plt.bar(sub_set_tiros['Squad'], sub_set_tiros['SoT'], color = 'red')
plt.title('Tiros por equuipo')
plt.ylabel('SoT')
```
